##**Create connection to database**


###**import libraries and load database**

```python


import sqlite3
import pandas as pd
import matplotlib.pyplot as plt


conn = sqlite3.connect('AUS_energy_sqlite3.db')
cursor = conn.cursor()
```

```python
sql = '''
SELECT *
FROM AUS_energy;
'''
AUS_energy_df = pd.read_sql_query(sql,conn)
print(AUS_energy_df)
```

###**"Data Cleansing: Replacing Empty Values with Zero in Energy Sources"**

```python
#To clean up the by replace blank fill with '0'
AUS_energy_df['Coal (PJ)'] = AUS_energy_df['Coal (PJ)'].replace('',0)
AUS_energy_df['Oil (PJ)'] = AUS_energy_df['Oil (PJ)'].replace('',0)
AUS_energy_df['Gas (PJ)'] = AUS_energy_df['Gas (PJ)'].replace('',0)
AUS_energy_df['Renewables (PJ)'] = AUS_energy_df['Renewables (PJ)'].replace('',0)

AUS_energy_df
```

##**Group A, Q:2 Computing and visualizing the total energy consumption by fuel type**









This question aim to provide the total energy consumption by fuel type Of Australia.

To analyze this we need to filter required data from database.

```python
# Define a list of fuel types to be included in the filtered DataFrame
fuel_type = ['Coal (PJ)', 'Oil (PJ)', 'Gas (PJ)', 'Renewables (PJ)']

# Filter the DataFrame to include only the columns with fuel types in the list
filtered_df = AUS_energy_df[fuel_type]

# Display the filtered DataFrame
print(filtered_df)
```

```python

# @title Overview of Sample Data { run: "auto" }
FuelType = "All fuel" # @param ["Coal (PJ)", "Oil (PJ)", "Gas (PJ)", "Renewables (PJ)", "All fuel"]

#To category fuel types
fuel_types = ['Coal (PJ)', 'Oil (PJ)', 'Gas (PJ)', 'Renewables (PJ)']

#Calculating the sum of energy consumption for each fuel type
coal_sum = filtered_df['Coal (PJ)'].sum()
oil_sum = filtered_df['Oil (PJ)'].sum()
gas_sum = filtered_df['Gas (PJ)'].sum()
renewables_sum = filtered_df['Renewables (PJ)'].sum()

# Creating a dictionary to map fuel types to their respective sums
fuel_sums = {
    'Coal (PJ)': coal_sum,
    'Oil (PJ)': oil_sum,
    'Gas (PJ)': gas_sum,
    'Renewables (PJ)': renewables_sum
}
#Defining colors for each fuel type for the bar plot
colors = ['blue', 'orange', 'green', 'pink']
# Check selected FuelType and display corresponding value
if FuelType in fuel_sums:
    selected_sum = fuel_sums[FuelType]
    #Plotting the bar graph for the selected fuel type
    plt.figure(figsize=(3, 8))
    bars =plt.bar([FuelType], [selected_sum], color=colors[fuel_types.index(FuelType)], width=0.1)
    plt.title(f'Total Energy Consumption for {FuelType}')
    plt.xlabel('Fuel Type')
    plt.ylabel('Total Energy Consumption (PJ)')
    plt.bar_label(bars, fmt='%.2f')  # Annotate the bar with its value
    plt.show()
else:
    #If the FuelType is not valid, display all fuel types
    energy_consumption = [fuel_sums[fuel] for fuel in fuel_types]
    #Plotting the bar graph for all fuel types
    plt.figure(figsize=(8, 6))
    bars = plt.bar(fuel_types, energy_consumption, color=['blue', 'orange', 'green', 'pink'])
    plt.title('Total Energy Consumption by Fuel Type')
    plt.xlabel('Fuel Type')
    plt.ylabel('Total Energy Consumption (PJ)')
    plt.bar_label(bars, fmt='%.2f')  # Annotate each bar with its value
    plt.show()
```

**ANALYSIS**: Over the years, the largest energy consumptions by all residents in Australia is oil. The chart shows how much energy is used from different types of fuel: coal, oil, gas and renewables. The most used energy source is oil, with 31,130 PJ. Next is coal with 26,946.90 PJ, followed by gas at 20,049.60 PJ. The lowest use is for renewable energy sources, with only 5,054.90 PJ of power available.  It means that, in fact, the majority of our energy is derived from oil, coal and gas which are not renewable sources. We still rely heavily on nonrenewable fuels even though there is a push to use more renewable energy for the health of our planet. There is a need to do more to increase the use of renewable energy sources and to make them more sustainable, given the low number of renewable energy sources.

##**Group A, Q:3 Analyse and plot the total energy consumption fuel types for each year**

This question aim to provide the total energy consumption  fuel type for each year Of Australia.

To analyze this we need to filter required data(Year and all fuel type) from database.

```python
# Define a list of fuel types to be included in the filtered DataFrame
element = ['Year','Coal (PJ)', 'Oil (PJ)', 'Gas (PJ)', 'Renewables (PJ)']

# Filter the DataFrame to include only the columns with fuel types in the list
filtered_df = AUS_energy_df[element]
yearly_df = filtered_df.groupby('Year').sum()

# Display the filtered DataFrame
print(yearly_df)
```

```python
# @title Overview of sample Data
FuelType = "All Fuel type" # @param ["Coal (PJ)", "Oil (PJ)", "Gas (PJ)", "Renewables (PJ)", "All Fuel type"] {allow-input: true}
Year = "All Year" # @param ["2008-09", "2009-10", "2010-11", "2011-12", "2012-13", "2013-14", "2014-15", "2015-16", "2016-17", "2017-18", "2018-19", "2019-20", "2020-21", "2021-22", "All Year"] {allow-input: true}

# Defining fuel types and colors for the bar plot
fuel_types = ['Coal (PJ)', 'Oil (PJ)', 'Gas (PJ)', 'Renewables (PJ)']
colors = ['blue', 'orange', 'green', 'pink']


if FuelType in fuel_types and Year in yearly_df.index:
    # Plotting the bar graph for the selected fuel type and year
    plt.figure(figsize=(4, 6))
    plt.bar(FuelType, yearly_df.loc[Year, FuelType], color=colors[fuel_types.index(FuelType)])
    plt.title(f'Energy Consumption for {FuelType} in {Year}')
    plt.xlabel('Fuel Type')
    plt.ylabel('Energy Consumption (PJ)')
    plt.grid(True)
    plt.show()
elif FuelType not in fuel_types or Year not in yearly_df.index:
    # Plotting the bar graph for all years
    plt.figure(figsize=(10, 6))
    for fuel_type, color in zip(fuel_types, colors):
        plt.plot(yearly_df.index, yearly_df[fuel_type], label=fuel_type, color=color, marker='o')
       # Annotating each data point with its value
        for x, y in zip(yearly_df.index, yearly_df[fuel_type]):
            plt.text(x, y, f'{y:.2f}', ha='center', va='bottom')
    plt.title('Energy Consumption for All Fuel Type')
    plt.xlabel('Year')
    plt.ylabel('Energy Consumption (PJ)')
    plt.xticks(rotation=45)
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()

















```

**ANALYSIS**:  The graph illustrates energy consumption trends from 2008-09 to 2021-22 across various fuel types: coal, oil, gas, and renewables, measured in petajoules (PJ). In particular, the use of coal continued to decrease steadily from 2350.90 million tonnes per annum to 1586.30 million tonnes per annum as a result of an upward trend away from coal in view of its impact on the environment. Over the years, oil consumption has fluctuated but is currently declining in line with global efforts to limit reliance on coal.
 Gas consumption has been increasing steadily, which may indicate a shift to more environmentally friendly fuels although it continues to contribute to GHG emissions. In contrast, renewable energy consumption continued to grow steadily from 240.00 PJh to 512.80 PJh, indicating a positive trend for sustainability. Policies to promote clean energy and technological progress influence this change.
 In the future, it is crucial to continue investing in renewables and strategies for a just transition away from fossil fuels if we are to remain sustainable. This trend highlights the importance of switching to renewable energy sources for a cleaner, more sustainable future, in line with global initiatives to achieve carbon neutrality and mitigate climate change.

```python
# @title Write a funtion to plot a Graph for visualization
FuelType = "Total fuel type" # @param ["Coal (PJ)", "Oil (PJ)", "Gas (PJ)", "Renewables (PJ)", "Total fuel type"] {allow-input: true}
State = "All State" # @param ["WA", "NSW", "VIC", "TAS", "SA", "QLD", "NT", "All State"]
 #Defining fuel types and colors for the bar plot
fuel_types = ['Coal (PJ)', 'Oil (PJ)', 'Gas (PJ)', 'Renewables (PJ)']
if FuelType in fuel_types and State in filtered_df['State'].unique():
    plt.figure(figsize=(4, 8))
    plt.bar(grouped_by_state[grouped_by_state['State'] == State]['State'], grouped_by_state[grouped_by_state['State'] == State][FuelType], color='skyblue')
    plt.title(f'Total {FuelType} Consumption in State {State}')
    plt.xlabel('State')
    plt.ylabel(f'Total {FuelType} Consumption (PJ)')
    plt.xticks(rotation=45)
    plt.grid(axis='y')
    plt.tight_layout()
    plt.show()
else:
    # If both the State and FuelType are not valid, display total energy consumption for each state for all fuel types
    plt.figure(figsize=(12, 8))
    for state in filtered_df['State'].unique():
        plt.bar(state, grouped_by_state[grouped_by_state['State'] == state]['Total_Energy_PJ'], label=state)

    plt.title('Total Energy Consumption for Each State for All Fuel Types')
    plt.xlabel('State')
    plt.ylabel('Total Energy Consumption (PJ)')
    plt.xticks(rotation=45)
    plt.legend()
    plt.grid(axis='y')
    plt.tight_layout()
    plt.show()
```

##**Group C, Q:08 Analysis the energy intensity, defined as energy consumption per unit of GSP, for each state annually and chart these figures**

Here we are trying to analyse the total energy intensity defined as energy consuption per unit of GSP, foe each state annually in Australia.
 For that With the help of following code we can extract data from database about all 4 type of fuel type  and GSP($millon) according to state and year.

 Note: From instruction  we were given 1PJ= 1000GJ So Calculation done according to this. where as https://www.convert-measurement-units.com/conversion-calculator.php this coverter provide 1PJ = 100000GJ

```python
# Define a list of fuel types to be included in the filtered DataFrame
element2 = [ 'Year','State','Coal (PJ)', 'Oil (PJ)', 'Gas (PJ)', 'Renewables (PJ)', 'GSP ($ million)']

# Filter the DataFrame to include only the columns with fuel types in the list
filtered_df = AUS_energy_df[element2]

# Display the filtered DataFrame
print(filtered_df)
```

```python

# Calculate the total energy consumption in petajoules (PJ) by summing up the energy consumption
filtered_df['Total_Energy_PJ'] = filtered_df['Coal (PJ)'] +filtered_df['Oil (PJ)'] + filtered_df['Gas (PJ)'] + filtered_df['Renewables (PJ)']
# Convert total energy consumption to GJ
filtered_df['Total_Energy_GJ'] = filtered_df['Total_Energy_PJ'] * 1_000
# Calculate energy intensity in GJ per $ million of GSP
filtered_df['Energy_Intensity'] = filtered_df['Total_Energy_GJ'] / filtered_df['GSP ($ million)']
print(filtered_df)
```

```python
grouped_by_state_year = filtered_df.groupby(['State', 'Year']).mean()
print(grouped_by_state_year)
```
